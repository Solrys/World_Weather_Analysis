```python
def randomCityGen(size):
    lats = np.random.uniform(-90.000, 90.000, size=2000)
    lngs = np.random.uniform(-180.000, 180.000, size=2000)
    result = zip(lats, lngs)
    #print (result)
    coordinates = list(result)
    # print coordinates
    cities = [] 
    for coordinate in coordinates:
        cities.append(citipy.nearest_city(coordinate[0], coordinate[1]).city_name)

#-------------------------------------------------------------------------
    return cities
    
```
